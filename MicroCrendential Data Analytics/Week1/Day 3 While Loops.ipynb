```python
prompt = "Input a message. I will repeat it until you type 'quit'"
message=""

while message != 'quit':
  message = input(prompt)
  print(message)
```

```python
#Flags with while loops
prompt = "Input a message. I will repeat it until you type 'quit'"
message = ""

active = True
while active: 
  message = input(prompt)
  
  if message = 'quit':
    active = False 

  else: 
    print(message)
```
