##String Indexing
We know strings are a sequence, which means Python can use indexes to call parts of the sequences. Let's learn how this works. In Python, we use brackets [] after an object to call its index. We should also note that indexing starts at 0 for Python. Let's create a new object called s and the walk through a few example of indexing. 

```python
s[0] 
```

```python
s[6]
```
