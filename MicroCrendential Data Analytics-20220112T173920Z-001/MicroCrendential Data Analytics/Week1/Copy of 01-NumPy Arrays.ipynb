## Using NumPy

Once you've installed NumPy you can import it as a library:

```python
import numpy as np
```

Numpy has many built-in functions and capabilities. We will not have the time to cover them all; we will, however, focus on some of the most important topics such as vectors, arrays, matrices, and number generation. Let's start by discussing arrays.

## Numpy Arrays

NumPy arrays are the main objects that we will use in this class. 

Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-dimensional arrays and matrices are 2-dimensional (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

### Creating NumPy Arrays From a Python List

We can create an array by directly converting a list or list of lists:

```python
#creating a list
my_list = [1,2,3]
my_list
```

```python
#converting list into a array
np.array(my_list)
```

```python
#matrix with 3 arrays within it 
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
my_matrix
```

```python
#outputting the matrix 
np.array(my_matrix)
```

## Built-in Methods

There are lots of built-in ways to generate Arrays

### arange

Return evenly spaced values within a given interval.

```python
#list out numbers from 0-19
np.arange(0,20)
```

```python
#counts by 2 and list out number from 0-118
np.arange(0,110,2)
```

### zeros and ones

Generate arrays of zeros or ones

```python
#outputs a vector of 0, 5times
np.zeros(5)
```

```python
#outputs a matrix of 0, 15x15
np.zeros((15,15))
```

```python
#outputs a vector of ones, 30 times
np.ones(30)
```

```python
#Outputs a matrix of ones, 30x30
np.ones((30,30))
```

### linspace
Return evenly spaced numbers over a specified interval.

```python
#returns 5 evenly spaced numbers between 0-100
np.linspace(0,100,5)
```

```python
#returns 50 evenly spaced numbers between 0-10
np.linspace(0,10,50)
```

## eye

Creates an identity matrix

```python
#returns a 6x6 matrix
np.eye(6)
```

## Random 

Numpy also has lots of ways to create random number arrays:

### rand
Create an array of the given shape and populate it with
random samples from a uniform distribution
over ``[0, 1)``.

```python
#generates 5 random number over the interval [0,1]
np.random.rand(5)
```

```python
#generates a 10x10 matrix that generates 10 random number over the interval [0,1]
np.random.rand(10,10)
```

### randn

Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

```python
#generates 20 random numbers from the "standard normal" distribution
np.random.randn(20)
```

```python
#generates 5 random numbers from the "standard normal" distribution in a 5x5 matrix
np.random.randn(5,5)
```

### randint
Return random integers from `low` (inclusive) to `high` (exclusive).

```python
#returns a random integer between 1 and 500
np.random.randint(1,500)
```

```python
#returns 100 random integer from 1-999
np.random.randint(1,1000,100)
```
